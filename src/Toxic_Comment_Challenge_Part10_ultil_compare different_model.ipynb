```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


plt.rcParams['figure.dpi']= 150
sns.set(style="whitegrid")
```

```python
nb = pd.read_csv("naive_bayes_compare_param.csv")
```

```python
nb = nb.where(nb["param"] == "cleaned").dropna()
```

```python
nb
```

```python
nb.columns = ["value", "score_name", "category", "model"]
```

```python
nb["model"] = "NB"
```

```python
nb
```

```python
boost = pd.read_csv("boosting.csv")
decisionTree = pd.read_csv("decisionTree_cross_validation.csv")
randomForest= pd.read_csv("randomForest.csv")
decisionTree.columns = ["value", "score_name", "category", "model"]
decisionTree["model"] = decisionTree["model"].apply(lambda x: "decisionTree")
compare = decisionTree.append(randomForest, ignore_index=True)
compare = compare.append(boost, ignore_index=True)
compare["model"] = compare["model"].apply(lambda x: "DT" if x == "decisionTree" else x)
compare["model"] = compare["model"].apply(lambda x: "RF" if x == "randomForest" else x)
compare["model"] = compare["model"].apply(lambda x: "AB" if x == "boosting" else x)

```

```python
compare = compare.append(nb, ignore_index=True)
```

```python
compare
```

```python
svm = pd.read_csv("SVM.csv")
```

```python
compare = compare.append(svm, ignore_index=True)
```

```python
lg = pd.read_csv("logisticRegression.csv")
```

```python
compare = compare.append(lg, ignore_index=True)
```

```python
compare["model"] = compare["model"].apply(lambda x: "LG" if x == "logisticRegression" else x)
```

```python
plt.rcParams['figure.dpi']= 150
ax = sns.catplot(x="model", y="value", col="category", data=compare, col_wrap=3, kind="box", height=4, aspect=.7);
```

```python
dt = compare.where(compare["model"] == "DT").dropna()
rf = compare.where(compare["model"] == "RF").dropna()
ab = compare.where(compare["model"] == "AB").dropna()
nb = compare.where(compare["model"] == "NB").dropna()
svm = compare.where(compare["model"] == "SVM").dropna()
lg = compare.where(compare["model"] == "LG").dropna()
```

```python
result=pd.DataFrame()
result = result.append(dt, ignore_index=True)
result = result.append(nb, ignore_index=True)
result = result.append(rf, ignore_index=True)
result = result.append(ab, ignore_index=True)
result = result.append(lg, ignore_index=True)
result = result.append(svm, ignore_index=True)
```

```python
plt.rcParams['figure.dpi']= 150
ax = sns.catplot(x="model", y="value", col="category", data=compare, kind="box", height=4, aspect=.7);
```
